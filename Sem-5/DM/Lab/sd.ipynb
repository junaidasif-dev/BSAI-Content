```python
from scipy.stats import norm
import statistics
```

```python
df.min(), df.max()
```

```python
x_axis = np.arange(df['single_price'].min(), df['single_price'].max())
```

```python
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
print(mean)
print(sd)
```

```python
plt.plot(x_axis, norm.pdf(x_axis, mean, sd)) #pdf = probability density function
plt.show()
```

```python
x_axis = np.arange(df['amount'].min(), df['amount'].max())
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis.tolist())
plt.plot(x_axis, norm.pdf(x_axis, mean, sd)) #pdf = probability density function
plt.show()
```
